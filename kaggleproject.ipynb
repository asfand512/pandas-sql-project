```python
# extract data from zip file 
import zipfile
zip_ref=zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract file to dir 
zip_ref.close() #close file
```

```python
# read data from the file and remove null values
import pandas as pd
#na_values will change all garbage values to nan
df=pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#rename the columns -- make them lower case and replace space with under score
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.head(5)
```

```python
# create new columns discount, sale price and profit
df['discount']= df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
# convert order_date from object datatype to datetime

df['order_date']=pd.to_datetime(df['order_date'],format="mixed")
# df.dtypes
```

```python
# drop the unnessesary columns

#this will return error after runnnig one time because the columns will not b there
# df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace=True)


# List of columns to drop
columns_to_drop = ['list_price', 'cost_price', 'discount_percent']

# Filter out the columns that are not in the DataFrame
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Drop the existing columns
df.drop(columns=existing_columns_to_drop, inplace=True)

# Display the DataFrame
df
```
